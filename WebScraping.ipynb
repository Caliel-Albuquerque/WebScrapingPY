```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt

# Fazendo a requisição para o site do Globo Esporte
url = 'https://ge.globo.com/'
response = requests.get(url)

# Verificando se a requisição foi bem sucedida
if response.status_code == 200:
    # Criando um objeto BeautifulSoup com o conteúdo da página
    soup = BeautifulSoup(response.text, 'html.parser')

    # Encontrando os elementos que contém as informações desejadas
    # Neste exemplo, vamos pegar as categorias das notícias principais
    news_categories = soup.find_all('span', class_='feed-post-metadata-section')

    # Extraindo as categorias das notícias
    categories = [category.text for category in news_categories]

    # Contando a frequência de cada categoria
    category_counts = {}
    for category in categories:
        if category not in category_counts:
            category_counts[category] = 0
        category_counts[category] += 1

    # Gerando uma visualização com as categorias das notícias
    plt.barh(range(len(category_counts)), list(category_counts.values()))
    plt.yticks(range(len(category_counts)), list(category_counts.keys()))
    plt.xlabel('Número de notícias')
    plt.title('Categorias das notícias principais do site do Globo Esporte')
    plt.show()
else:
    print('Erro ao fazer a requisição para o site do Globo Esporte')
```
